## Cell Phone Data

```python
import pandas as pd
df = pd.read_csv('cell_phones_total.csv')
df
```

We create a function to convert the string numeric values to floats and apply it to all columns except first.

```python
def convert_to_number(value):
    str_numbers = {'k': 1000, 'M': 1000000, 'B': 1000000000}
    if isinstance(value, str):
        if value[-1] in str_numbers:
            number = float(value[:-1])
            return number * str_numbers[value[-1]]
        else:
            return float(value)
    return value
df.iloc[:,1:] = df.iloc[:,1:].map(convert_to_number)
df
```

Then we want to fill all nan values with zero, along the rows, until we reach a value larger than zero indicating that recordings has begun for the given country.

```python
def first_fill(row):
    filled = []
    larger_than_zero = False
    for value in row:
        if not larger_than_zero:
            if pd.isna(value) or value == 0:
                filled.append(0)
            else:
                larger_than_zero = True
                filled.append(value)
        else:
            filled.append(value)
    return filled
df.iloc[:,1:] = df.iloc[:,1:].apply(first_fill, axis=1, result_type='expand')
df
```

Now we fill in the rest of the missing values by using the value before as an estimate for the missing value.

```python
def second_fill(row):
    filled = []
    for value in row:
        if pd.isna(value) and filled:
            filled.append(filled[-1])
        else:
            filled.append(value)
    return filled
df.iloc[:, 1:] = df.iloc[:, 1:].apply(second_fill, axis=1, result_type='expand')
df
```

Lastly we reverse the order of year columns and display the numbers in the format requested.

```python
year_columns = sorted([col for col in df.columns if col.isdigit()], reverse=True)
columns_order = ['iso-3'] + year_columns
df = df[columns_order]
pd.options.display.float_format = '{:.2e}'.format
df
```
