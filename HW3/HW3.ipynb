## IRIS data

We start of by importing pandas, seaborn and matplotlib in order to process and visualize the data. We import the csv file and in order to get a good overview of the material we print the columns.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('IRIS.csv')

print(df.columns)
```

We can see that the columns are straight forward so we give the columns of the data frame as arguements to the sns plot function directly and choose a dark grid layout. In order to plot the figures side by side we use the subplots function in plt which plots each figure on a grid, in this case with one row and 2 columns. In order to get the colors we want we customize a palette using hex codes. We then proceed to provide sns with the necessary arguments.

```python
sns.set_style('darkgrid')
custom_palette = custom_palette = ["#5a7db5", "#4ca648", "#c04b4b"]
sns.set_palette(sns.color_palette(custom_palette))

fig, axs = plt.subplots(1, 2, figsize=(12, 6))
sns.scatterplot(data=df,  x='sepal_length', y='petal_length', hue='species', ax=axs[0])
axs[0].set_title('Relationship between sepal- and petal length')

sns.scatterplot(data=df,  x='sepal_width', y='petal_width', hue='species', ax=axs[1])
axs[1].set_title('Relationship between sepal- and petal width')
plt.show()
```

Based on the left figure it seems that for Iris Setosa increasing sepal length does not significantly affect petal length. As for Iris Versicolor and Iris Virginica we do see a positive correlation between the two dimensions. The same observation seems to be true when we study the figure on the right. However the correlation between sepal- and petal width for Iris Versicolor and Iris Virginica seems to be, not as pronounced as the dimensions pertaining to length.

In order to create a visually appealing plot and clear plot we again use the subplot function of plt. Due to the crowded space of four plots in the same frame we remove label of the x-axis as its contents is clear soley by providing the tick labels for each subplot. We costumize the fliers of the box plots to make them fit better and reuce all font- and label sizes. The y-labels are also modified to look better.

```python
fig, axs = plt.subplots(2, 2)
sns.boxplot(data=df, x='species', y='sepal_width', hue='species',
            flierprops={'marker' : 'o',
                        'markerfacecolor' : 'none',
                        'markeredgecolor' : 'black',
                        'markeredgewidth' : 0.5,
                        'markersize' : 8}, 
            linewidth=0.5, ax=axs[0,0])
axs[0,0].set_xlabel('')
axs[0,0].set_ylabel('Sepal width', fontsize=7)
axs[0,0].tick_params(axis='x', labelsize=7)
axs[0,0].tick_params(axis='y', labelsize=7)

sns.boxplot(data=df, x='species', y='sepal_length', hue='species',
            flierprops={'marker' : 'o',
                        'markerfacecolor' : 'none',
                        'markeredgecolor' : 'black',
                        'markeredgewidth' : 0.5,
                        'markersize' : 8}, 
            linewidth=0.5, ax=axs[0,1])
axs[0,1].set_xlabel('')
axs[0,1].set_ylabel('Sepal length', fontsize=7)
axs[0,1].tick_params(axis='x', labelsize=7)
axs[0,1].tick_params(axis='y', labelsize=7)

sns.boxplot(data=df, x='species', y='petal_width', hue='species',
            flierprops={'marker' : 'o',
                        'markerfacecolor' : 'none',
                        'markeredgecolor' : 'black',
                        'markeredgewidth' : 0.5,
                        'markersize' : 8}, 
            linewidth=0.5, ax=axs[1,0])
axs[1,0].set_xlabel('')
axs[1,0].set_ylabel('Petal width', fontsize=7)
axs[1,0].tick_params(axis='x', labelsize=7)
axs[1,0].tick_params(axis='y', labelsize=7)

sns.boxplot(data=df, x='species', y='petal_length', hue='species',
            flierprops={'marker' : 'o',
                        'markerfacecolor' : 'none',
                        'markeredgecolor' : 'black',
                        'markeredgewidth' : 0.5,
                        'markersize' : 8,}, 
            linewidth=0.5, ax=axs[1,1])
axs[1,1].set_xlabel('')
axs[1,1].set_ylabel('Petal length', fontsize=7)
axs[1,1].tick_params(axis='x', labelsize=7)
axs[1,1].tick_params(axis='y', labelsize=7)

plt.tight_layout()
plt.show()
```

As for the distribution of sepal- and petal dimensions we can start of by noting that regarding both length and width dimensions across all species seems to be approximately equally distributed around the median. The most striking anomaly visually is the discrepancy in relative size of interquartile range between Iris Setosa and the remaining species regarding sepal width where we find the only example of Iris Setosa having an interquartile range larger than both of its counterparts. In all other dimensions we find that the interquartile range of Iris Setosa is smaller or ever much smaller than those of Iris Versicolor and Iris Virginicia.

## Birdwatching

First of we need to operationalize the variable "prevalence". Hence we first decide that prevalence is the total number of individuals of a species present in the park during a time intervall. We then open, read the csv-file and print the columns.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('artportalen.csv')
print(df.columns)
```

Based on the columns we are interested in "Antal" and also in the number of reported sightings. Assuming number of "reporters" are somewhat evenly spread out over a month we operationalize prevalence as follows. Prevalence := \sum (#antal) over all sightings of the species.

We need to clean the data so we drop potential duplicates and rows with NaN values in the 'Antal' column. 

```python
df.drop_duplicates(inplace=True)
df.dropna(subset=['Antal'], inplace=True)
```

We then check the data types of inside 'Antal'.

```python
print(df['Antal'].apply(type).unique())
```

We see that its all strings and so we look for any non numerical string values being used inside 'Antal'.

```python
non_numerics = df['Antal'][df['Antal'].str.isnumeric() == False].unique()
print(non_numerics)
```

The fact that its noted does not help us so we remove those values. However if its not found then that would equal zero so we replace all instances of not found with zero.

```python
df.drop(df[df['Antal'] == 'noterad'].index, inplace=True)
df.replace('Ej återfunnen','0', inplace=True)
```

Now we want to calculate prevalence and so we convert the values in 'Antal' to numeric values and sum the number of of individuals over all sightings for each species. Then we sort the species in descending order. 

```python
df['Antal'] = pd.to_numeric(df['Antal'], errors='coerce')
total_individuals = df.groupby('Artnamn')['Antal'].sum().sort_values(ascending=False).reset_index()

print(total_individuals)
```

It seems the most prevalent species are Gröniska, Sothöna, Gräsand, Storskrake and Vitkindad gås. 

In order to find the distribution of the top 3 most prevalent species we start of by creating a list with the top 3. We then use the 'Startdatum' column to create a new column representing the month of sighting.

```python
top_3_list = [x for x in total_individuals.head(3)['Artnamn']]
df['Månad'] = pd.to_datetime(df['Startdatum'], errors='coerce').dt.month
filtered_df = df[df['Artnamn'].isin(top_3_list)]
monthly_dist = filtered_df.groupby(['Artnamn', 'Månad'])['Antal'].sum().reset_index()
print(monthly_dist)
```

We can see that for all top prevalent species there is a decline trend in numbers from january to june. We plot the monthly distribution with a lineplot for each species which visually confirms the numbers.

```python
plt.figure()
sns.lineplot(data=monthly_dist, x='Månad', y='Antal', hue='Artnamn')
plt.title('Distribution of prevalent species over months')
plt.show()
```

In order to find the rarest species we use the variable total_individuals and reverse the order of sorting.

```python
print(total_individuals.sort_values(by='Antal', ascending=True))
```

We see there are a couple of species with merely one observation and so we select and print all those species.

```python
rarest_species = total_individuals[total_individuals['Antal'] == 1]
print(rarest_species)
```

### Three questions

Our three questions are a) Are there species which only reside in specific areas within the national park? b) Which are the most stable species over time? c) Which are the species that coexist the least in the park?

In order to answer the first question we split the park in to the municipalities represented in the park and get all unique species.

```python
artnamn = df['Artnamn'].unique()
kommuner = df['Kommun'].unique()
```

We then create a data frame with all combinations of species and municipalities using list comprehension

```python
all_combinations = pd.DataFrame([(a, k) for a in artnamn for k in kommuner], columns=['Artnamn', 'Kommun'])
print(all_combinations)
```

We then group based on species and municipality and merge with the combinations df creating a new df showing the sumed number for each species and municipality.

```python
number_municip = df.groupby(['Artnamn', 'Kommun'], as_index=False)['Antal'].sum()
single_municip_spec = all_combinations.merge(number_municip, on=['Artnamn', 'Kommun'], how='left').fillna({'Antal': 0})
sms = single_municip_spec
print(sms)
```

We calculate the total number of each species and add it as a column. We then calculate the percentage of total for each species and municipality.

```python
total_species = sms.groupby('Artnamn', as_index=False)['Antal'].sum()
total_species.rename(columns={'Antal': 'Total_Antal'}, inplace=True)
sms = sms.merge(total_species, on='Artnamn')
sms['Percentage'] = (sms['Antal'] / sms['Total_Antal']) * 100
sms = sms.sort_values(by=['Artnamn', 'Percentage'], ascending=[True, False])
print(sms)
```

We select those species where the percentage is a 100 percent.

```python
sms = sms[sms['Percentage'] == 100.0]
print(sms)
```

To get a sense of significance we calculate the number of observations for each "single municipality species". A large number of observations will give strength to the conclusion that the species only resides in a specific municipality.

```python
observation_counts = df['Artnamn'].value_counts().reset_index()
observation_counts.columns = ['Artnamn', 'Observations']
sms = sms.merge(observation_counts, on='Artnamn')
top_10_species = sms.sort_values(by='Observations', ascending=False).head(10)
print(top_10_species)
```

Of these 10 we choose the top 5 where the minimum number of observations is 31. We are then confident that Brandkronad kungsfågel, Enkelbeckasin, Svarthakedopping, Bläsand and Stenskvätta only resides in specific areas of the park. 

In order to find the most stable species we reuse the concept of monthly_dist but for all species and pivot the df.

```python
total_monthly_dist = df.groupby(['Artnamn', 'Månad'])['Antal'].sum().reset_index()
tmd = total_monthly_dist
tmd = tmd.pivot_table(index='Artnamn', columns='Månad', values='Antal', 
                                               aggfunc='first', fill_value=0).reset_index()
tmd.drop(columns=12, inplace=True)
tmd.columns.name = None
print(tmd)
```

We measue stability as by standard deviation across the months. We are also interested in the "stable and present" species so we select only those present during all months.

```python
species_std = tmd.iloc[:, 1:].std(axis=1)
tmd['std'] = species_std
tmd = tmd[(tmd.iloc[:, 1:] != 0).all(axis=1)]
print(tmd)
```

We now choose the 10 species with the lowest stds which are also by our previous definition the most stable species.

```python
stable_top_ten = tmd.sort_values(by='std').head(10)
print(stable_top_ten)
```

We now want to see which species do not coexist in the park and so we measure the correlations of species across the months by pivoting the original total_monthly_dist but this time with species as columns.

```python
tmd = total_monthly_dist.pivot_table(index='Månad', columns='Artnamn', values='Antal', 
                                               aggfunc='first', fill_value=0).reset_index()
tmd.drop(index=6, inplace=True)
tmd.columns.name = None
print(tmd)
```
