## Joining data

We read the csv file of the population data and view its contents.

```python
import pandas as pd
population_data = pd.read_csv('pop_data.csv')
popd = population_data
popd
```

The column 'Unnamed: 0' is equivalent with the indices so we remove it. At first we replaced the NaN values with 0 but when we encountered the task further down, setting the NaN values to 0 would give a false sense of the rate of growth. Hence we let the NaN values be NaN.

```python
popd = popd.drop(columns='Unnamed: 0')
# popd = popd.fillna(0)
print(popd.isna().any().any())
popd
```

We use melt to get the desired format and rename the columns.

```python
popd = pd.melt(popd, id_vars='iso-3')
popd.columns = ['iso-3', 'year', 'population']
popd
```

We read the cell phone data, melt it and merge the data frames.

```python
cell_phone_clean = pd.read_csv('cell_phone_clean.csv')
cpc = cell_phone_clean
cpc = pd.melt(cpc, id_vars='iso-3')
cpc.columns = ['iso-3', 'year', 'n_cellphones']

merged = pd.merge(popd, cpc, how='outer')
merged = merged.sort_values(by=['year', 'iso-3'])
# merged = merged.fillna(0)
merged
```

In order to get the correct names for the countries we read the csv file country_data.

```python
country_data = pd.read_csv('country_data.csv')
cd = country_data
cd = cd[['alpha-3', 'name']]
merged = pd.merge(merged, cd, left_on='iso-3', right_on='alpha-3', how='left')
merged
```

We see that some abbreviations arent defined in the country data and so we replace the NaN values with the abbreviations for lack of better options. We then filter the df to contain only the columns of interest.

```python
merged['name'] = merged['name'].fillna(merged['iso-3'])
merged = merged[['name', 'year', 'n_cellphones', 'population']]
merged
```

We add a column based on the quotient of n_cellphones/population and filter the rows to contain only name, year, and cells per capita.

```python
merged.loc[:, 'cell_per_capita'] = merged['n_cellphones'] / merged['population']
merged = merged[['name', 'year', 'cell_per_capita']]
merged
```

We pivot the table placing the names on the x-axis and use the years as indices. We then order the columns based on the cell_per_capita for 2019. We then chose the years 2015-2019.

```python
top_2019 = merged.pivot(index='year', columns='name', values='cell_per_capita')
order_by_2019 = top_2019.loc['2019'].sort_values(ascending=False).index
top_2019 = top_2019.loc[:, order_by_2019]
top_2019 = top_2019.loc[['2015', '2016', '2017', '2018', '2019']]
top_2019
```

We now choose the top five countries and proceed to analyze their growth from the past 10 years. We see that 2020 is the most recent year but is lacking in data so we focus on 2010-2019.

```python
top_5 = merged.pivot(index='year', columns='name', values='cell_per_capita')
# print(merged['year'].unique())
top_5 = top_5 = top_5.loc['2010':'2019', 
['Hong Kong', 'United Arab Emirates', 'Seychelles', 'Montenegro', 'Thailand']]
top_5.sort_index()
top_5
```

We subtract the cell_per_capita number by the 2010 value for each country in order to get an equal starting point. Then we melt it in order to get it into the right format.

```python
standardized_growth = top_5.apply(lambda x: x - top_5.loc['2010'], axis=1)
growth = standardized_growth.reset_index()
growth = growth.melt(id_vars='year', var_name='country', value_name='value')
# growth
```

We import matplotlib.pyplot and seaborn and plot the figure.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
plt.figure()
sns.lineplot(data=growth, x='year', y='value', hue='country')
plt.xlabel('Year')
plt.ylabel('Cell phones per capita')
plt.title('Growth rate of cell phones per capita')
plt.show()
```

It seems Hong Kong has had the largest overall growth rate of the countries.

## SQL

We start of by importing sqlite3 and initializing a connection to the database. We then create a cursor.

```python
import sqlite3
con = sqlite3.connect('user_actions.db')
cur = con.cursor()
```

We now want to get an overview of the contents of the table and so we create a query and use PRAGMA.

```python
overview = ''' 
PRAGMA table_info('user_actions'); 
'''
res = cur.execute(overview)
res.fetchall()
```

We are interested in the usernames and so we select the distinct usernames in the database.

```python
usernames = ''' 
SELECT DISTINCT username 
FROM user_actions; 
'''
res = cur.execute(usernames)
print(res.fetchall())
```

We do the same thing regarding the different actions and then we select allunique users who have performed the signup.

```python
actions = ''' 
SELECT DISTINCT action 
FROM user_actions; '''
res = cur.execute(actions)
print(res.fetchall())
```

```python
signup_users = ''' 
SELECT DISTINCT username
FROM user_actions 
WHERE action = 'signup'; 
'''
res = cur.execute(signup_users)
print(res.fetchall())
```

We select user_id, username and count actions. We then group by username and to make it neat we order by user id.

```python
count_actions = '''
SELECT user_id, username, COUNT(*) action
FROM user_actions
GROUP BY username
ORDER BY user_id;
'''
res = cur.execute(count_actions)
res.fetchall()
```

In order to filter based soley on dates rather than dates and times we use the DATE method. We then filter out all actions but sign up and log in. We group on the users and dates and specify that at least 2 actions must be present in the group in order to ensure both the signup and login is in the group.

```python
sign_and_log = '''
SELECT user_id, username
FROM user_actions
WHERE action IN ('signup', 'login')
GROUP BY user_id, DATE(timestamp)
HAVING COUNT(DISTINCT action) >= 2
ORDER BY user_id;
'''
res = cur.execute(sign_and_log)
res.fetchall()
```

We then close the connection to the database.

```python
con.close()
```

## RegEx

We start by importing RegEx module and read the file. We then split the comments so that we can access each one separately and create a small function to adjust for the indexing.

```python
import re
with open('comments.txt', 'r') as file:
    text = file.read()
comments = text.splitlines()

def getcomment(i):
    return comments[i - 1]
print(text)
```

We create a regex so that the expression starts with #, followed some word described in alphanumeric characters. We then apply the search on comment 1.

```python
regex = r'\#[\w]*\b'
re.findall(regex, getcomment(1))
```

To find if a comment contains both #python and #programming we create a regex that checks wether both strings exist within the comment. 

```python
regex = r'(?=.*#python)(?=.*#programming)'
match = re.search(regex, text)
print(match)
```

We check if applying the regex to comment 2 gives a match.

```python
match = re.search(regex, getcomment(2))
print(match)
```

We see that it does not contain both expressions.

In order to find all unique hashtags we use the first regex we created and loop over all comments.

```python
regex = r'\#[\w]*\b'
hashtags = re.findall(regex, text)
unique_hashtags = list(set(hashtags))
print(unique_hashtags)
```
